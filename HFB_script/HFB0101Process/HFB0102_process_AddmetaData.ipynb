```python
BCR=pd.read_csv('/data5/maolp/Gaofeng_All_matrix/Allcount/All_scanpyData/BCRTCR/Gao_All_BCR.csv')
```

```python
# Create a dictionary from TCR and BCR 'barcodename' to their respective labels
tcr_dict = {index: 'TCR' for index in TCR['barcodename']}
bcr_dict = {index: 'BCR' for index in BCR['barcodename']}

# Add new column 'label' to adata.obs, initialize it with 'None'
adata.obs['TCRBCRlabel'] = 'None'

# Update 'label' for indices found in TCR and BCR
adata.obs['TCRBCRlabel'] = adata.obs.index.map(lambda x: tcr_dict.get(x, adata.obs.loc[x, 'TCRBCRlabel']))
adata.obs['TCRBCRlabel'] = adata.obs.index.map(lambda x: bcr_dict.get(x, adata.obs.loc[x, 'TCRBCRlabel']))

# If both TCR and BCR exist, label as 'Both'
adata.obs['TCRBCRlabel'] = adata.obs.index.map(lambda x: 'Both' if x in tcr_dict and x in bcr_dict else adata.obs.loc[x, 'TCRBCRlabel'])
```
